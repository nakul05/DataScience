```python

data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', np.nan, 'Eve'],
    'Age': [20, np.nan, 22, 21, 20],
    'Grade': ['A', 'B', np.nan, 'B', 'A'],
    'Marks': [85, 76, np.nan, 89, 92]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)

```

```python

#dropna(series +dataframe)--> how parameter works like or 

# by default  .dropna(how='any')  --> it removes all the rows that has any missing value in a column


print(df.dropna())

print('')

print(df.dropna(how='all'))
```

```python
#fill na --> fills missing value(series+df)

print(df.fillna(0))
print(' ')
print(df['Marks'].fillna(df['Marks'].mean()))
```

```python
#forward fill(ffill) and backward (bfill)
'''replaces nan value with previous or next values'''

print(df)
print(' ')
print(df['Name'].fillna(method='ffill'))
```

```python
import pandas as pd
import numpy as np

# Create a dictionary with student data, including some missing and duplicate values
data = {
    'Student_ID': [101, 102, 103, 104, 104],  # Duplicate Student_ID for 104
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'David'],  # Duplicate Name 'David'
    'Age': [20, np.nan, 22, 21, 21],  # Duplicate Age 21
    'Grade': ['A', 'B', np.nan, 'B', 'B'],  # Duplicate Grade 'B'
    'Marks': [85, 76, np.nan, 89, 89]  # Duplicate Marks 89
}

# Create DataFrame
students = pd.DataFrame(data)

print(students)
```

```python
#duplicated:

students.duplicated()

```

```python
#drop duplicates

print(students.drop_duplicates(keep='first'))
print('')
print(students.drop_duplicates(keep='last'))
```

```python
#finding last match played by virat kohii in delhi

# ipl['all_players'] = ipl['Team1Players']+ipl['Team2Players']
ipl[(ipl['City']=='Delhi' )  &(ipl['all+players'].str.contains('V Kohli'))].drop_duplicates(keep='first')


```

```python
#drop(series+dataframe)

print(students.drop(columns='Name')) #dropping columns


print(students.drop(index=[0,3]))   #dropping rows
```

```python
#apply:

# lets add 5 marks to each student 

students['Marks'].fillna(method='ffill',inplace = True)
students


students['Updated Marks'] = students['Marks'].apply(lambda x:x+5)
```

```python
students
```

```python
# Categorize Students by Score


students['Intellegence'] = students['Updated Marks'].apply(lambda x: 'High' if x>=80 else 'Medium' if x>=70 else 'Low' )
```
